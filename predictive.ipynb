```python
#IMPORTING LIBRARIES
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

```python
#LOADING THE CSV -- DESCRIPTIVE_DATA.CSV
predictive_df = pd.read_csv('predictive_data.csv')
predictive_df.head(3)
```

```python
predictive_df = predictive_df.drop(['potential_customer'
                           ], axis=1)
predictive_df['potential_customer'] = predictive_df['afirmation_1'].replace('Puede que entre ahora y los próximos 2 años me compre una bicicleta eléctrica.', '1')
predictive_df['potential_customer'] = predictive_df['potential_customer'].replace('Nunca he pensado, ni pensaré en comprarme una bicicleta eléctrica.', '0')
predictive_df['potential_customer'] = predictive_df['potential_customer'].replace('Después del aislamiento por el COVID hubiese sido un momento mejor que este para comprar una bicicleta eléctrica.', '0')
predictive_df['potential_customer'] = predictive_df['potential_customer'].replace('Ya tengo bicicleta eléctrica.', '0')
predictive_df['potential_customer'] = predictive_df['potential_customer'].astype(int)
display(predictive_df.dtypes)
predictive_df.head(3)
```

```python

predictive_df['not_potential_customer'] = predictive_df['afirmation_1'].replace('Nunca he pensado, ni pensaré en comprarme una bicicleta eléctrica.', '1')
predictive_df['not_potential_customer'] = predictive_df['not_potential_customer'].replace('Puede que entre ahora y los próximos 2 años me compre una bicicleta eléctrica.', '0')
predictive_df['not_potential_customer'] = predictive_df['not_potential_customer'].replace('Después del aislamiento por el COVID hubiese sido un momento mejor que este para comprar una bicicleta eléctrica.', '0')
predictive_df['not_potential_customer'] = predictive_df['not_potential_customer'].replace('Ya tengo bicicleta eléctrica.', '0')
predictive_df['not_potential_customer'] = predictive_df['not_potential_customer'].astype(int)
display(predictive_df.dtypes)
predictive_df.head(3)
```

```python
predictive_df = predictive_df.drop(['afirmation_1'
                           ], axis=1)
predictive_df['gender'] = predictive_df['gender'].str.replace('Género fluido','Man')
predictive_df['gender'] = predictive_df['gender'].str.replace('Queer','Man')
predictive_df['gender'] = predictive_df['gender'].str.replace('Trans','Man')
```

```python
# REPLACING AGAIN AGE BY AGE RANGE
valores_a_reemplazar = {
    22:'18-25 years',
   32: '30-35 years',
    65: 'Más de 55 years',
    40: '35-45 years',
    50:'45-55 years'
}
 # Reemplaza los valores en la columna 'edad'
predictive_df['age'] = predictive_df['age'].replace(valores_a_reemplazar)

predictive_df['age']
```
