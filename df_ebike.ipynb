```python
import pandas as pd


```

```python
# Leer el archivo CSV y crear el dataframe
df = pd.read_csv("E-BIKE (BICICLETA ELÉCTRICA).csv")
```

```python
df
```

```python
columnas =list(df.columns)
```

```python
for titulo in columnas:
    if 'Puntuación' in titulo:
        df = df.drop(titulo, axis=1)
        print(f'Se ha eliminado la columna: {titulo}')
print('Eliminación de columnnas terminado')
print('Estas son las columnas que quedan:')
print(df.columns)
        
```

```python
columnas =list(df.columns)
for titulo in columnas:
    if 'Comentarios' in titulo:
        df = df.drop(titulo, axis=1)
        print(f'Se ha eliminado la columna: {titulo}')
print('Eliminación de columnnas terminado')
print('Estas son las columnas que quedan:')
print(df.columns)
```

```python
df['Marca temporal'] = df['Marca temporal'].str.split(' ')
```

```python
for i in range(len(df)):
    df['Fecha'] = df['Marca temporal'][i][0]
    df['Hora'] = df['Marca temporal'][i][1]
```
